```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X.shape, y.shape
```

```python
# convert y to ints
y = np.array(y).astype('int64')
```

```python
def plot_roc(y_bool, y_probas_max):
    fpr, tpr, thresholds = roc_curve(y_bool, y_probas_max)
    fig = plt.figure(figsize=(20, 15))
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlabel('False Positive Rate', size=25)
    plt.ylabel('True Positive Rate', size=25)
    plt.title("ROC Curve")
    plt.show()
```

```python
def run_all_models(X, y, **kwargs):
    params = dict()
    if 'random_state' in kwargs:
        params['random_state'] = kwargs['random_state']
    else:
        params['random_state'] = None
    
    models = [RandomForestClassifier,
              MLPClassifier,
              #CategoricalNB,
              LogisticRegression,
              DummyClassifier]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
    print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
    
    for m in models:
        
        try:
            clf = m(random_state=params['random_state'])
        except TypeError:
            print(m)
            clf = m()
            
        print(clf)
        print("Fitting model to training data...")
        clf.fit(X_train, y_train)
        
        y_pred = clf.predict(X_test)
        y_probas = clf.predict_proba(X_test)
        y_probas_max = y_probas.max(1)
        y_bool = (y_test == y_pred).astype('int64')
        
        print(y_test)
        print(y_pred)
        m = confusion_matrix(y_test, y_pred)
        
        print("\n {} METRICS {} \n".format('*'*20, '*'*20))
        
        print("Accuracy: {}\n".format(accuracy_score(y_test, y_pred)))
        print("Per class TPR: {}\n".format(utils.get_tpr_per_class(m)))
        print("ROC Curve:\n")
        plot_roc(y_bool, y_probas_max)
        precision = precision_score(y_test, y_pred, average='weighted')
        recall = recall_score(y_test, y_pred, average='weighted')
        f1 = f1_score(y_test, y_pred, average='weighted')
        print("Per class precision:\n {}\n".format(precision_score(y_test, y_pred, average=None)))
        print("Weighted precision: {}\n".format(precision))
        print("Per class recall:\n {}\n".format(recall_score(y_test, y_pred, average=None)))
        print("Weighted recall: {}\n".format(recall))
        print("Per class F1-score:\n {}\n".format(f1_score(y_test, y_pred, average=None)))
        print("Weighted F1-score: {}\n".format(f1))
        print("Confusion matrix:\n {}\n".format(m))
        
        
        print("{} END OF MODEL RESULTS {}\n\n".format('*'*30, '*'*30))
```

### Metrics to gather

1. accuracy = TP (sum of matrix diagonal) / (sum of matrix) DONE
2. Per class: TPR = TP / (TP + FN) DONE
3. ROC curve 
4. precision (PPV) = TP / (TP + FP)
5. recall (sensitivity) = TP / (TP + FN)
6. F1-score = 2 * (precision x recall) / precision + recall
7. confusion matrix and plot
